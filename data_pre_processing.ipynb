```python
'''
一些函数
'''
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
def pre_processing(train_in):
  record=[]
  for i in train_in['发票状态']:
      if(i=='有效发票'):
          record.append(1)
      else:
          record.append(0)
  train_in['发票状态']=record 
  return train_in 
#汇总每个公司的数据 略慢
def sumfun(get_in,name,feature_num):#'E%d'%1
    '''
    输入：原始数据，当前企业名称，第几个特征
    输出：该企业该特征的总和
    '''
    record=[]
    for i in range(len(get_in)):
        if(get_in['企业代号'][i]==name):
            record.append(get_in.iloc[i,feature_num])
    return sum(record)
#汇总每一列的数据
def convert(name_list,train_in):
    '''
    输入：企业名称的列表，输入的表格
    输出：该表格汇总好的数据
    '''   
    train_in1=[]
    for name in name_list:
        train_in1.append(sumfun(train_in,name,1))
    train_in2=[]
    for name in name_list:
        train_in2.append(sumfun(train_in,name,2))
    train_in3=[]
    for name in name_list:
        train_in3.append(sumfun(train_in,name,3))
    train_in4=[]
    for name in name_list:
        train_in4.append(sumfun(train_in,name,4))    
    train_in1,train_in2,train_in3,train_in4=np.array(train_in1),np.array(train_in2),np.array(train_in3),np.array(train_in4)    
    return np.vstack((train_in1,train_in2,train_in3,train_in4))
```

```python
#处理训练集的数据，并保存 很慢 一个小时以内
train_in=pd.read_excel('训练集进项.xlsx',0)
train_in=pre_processing(train_in)     
name_list=train_in['企业代号'].drop_duplicates()
name_list=name_list.values
train_in=convert(name_list,train_in)
# 保存.npy文件
np.save("train_in.npy", train_in)
train_out=pd.read_excel('训练集销项.xlsx',0)
train_out=pre_processing(train_out)  
train_out=convert(name_list,train_out)
# 保存.npy文件
np.save("train_out.npy", train_out)
```

```python
#处理测试集的数据，并保存---很慢，2-3个小时
test_in=pd.read_excel('测试集进项.xlsx',0)
test_in=pre_processing(test_in)  
name_list=test_in['企业代号'].drop_duplicates()
name_list=name_list.values
test_in=convert(name_list,test_in)
print(test_in,test_in.shape)
# 保存.npy文件
np.save("test_in.npy", test_in)
test_out=pd.read_excel('测试集销项.xlsx',0)
test_out=pre_processing(test_out)  
test_out=convert(name_list,test_out)
# 保存.npy文件
np.save("test_out.npy", test_out)
```
